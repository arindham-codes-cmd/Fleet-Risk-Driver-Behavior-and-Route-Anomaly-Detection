```python
# brake events per km
df_trip['BRAKE_PER_KM'] = (df_trip['TOTAL_BRAKE_EVENTS'] /df_trip['TRIP_DISTANCE_KM'].replace(0, np.nan))
```

```python
# route deviation per km
df_trip['ROUTE_DEV_PER_KM'] = (df_trip['MAX_ROUTE_DEVIATION'] /df_trip['TRIP_DISTANCE_KM'].replace(0, np.nan))
```

```python
# hour of day
df_trip['TRIP_START_HOUR'] = df_trip['TRIP_START_AT'].dt.hour
```

```python
# bucket time of day
df_trip['TIME_OF_DAY'] = pd.cut(df_trip['TRIP_START_HOUR'],bins=[-1, 6, 12, 18, 24],labels=['night', 'morning', 'afternoon', 'evening'])
```

```python
feature_cols = [
    'TRIP_DURATION_MINUTES',
    'TRIP_DISTANCE_KM',
    'AVG_SPEED',
    'MAX_SPEED',
    'AVG_ACCELARATION',
    'TOTAL_BRAKE_EVENTS',
    'MAX_ROUTE_DEVIATION',
    'AVG_ROUTE_DEVIATION_SCORE',
    'BEHAVIORAL_CONSISTENCY_INDEX',
    'BRAKE_PER_KM',
    'ROUTE_DEV_PER_KM',
    'TRIP_START_HOUR'
]

X = df_trip[feature_cols]
```

```python
y = df_trip['ANY_ANOMALOUS_EVENT'].astype(int)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.3, stratify = y , random_state = 42)
```

```python
from xgboost import XGBClassifier
xg_model = XGBClassifier()
```

```python
xg_model.fit(X_train , y_train)
```

```python
y_pred = xg_model.predict(X_test)
```

```python
from sklearn.metrics import *
acc_score = accuracy_score(y_test , y_pred)
auc_score = roc_auc_score(y_test, y_pred)
print("Accuracy score of XGBClassifier model is :", acc_score*100)
print("AUC Score of XGBClassifier model is :", auc_score*100)
```

```python
# probabilities (positive class)
y_proba = xg_model.predict_proba(X_test)[:,1]

from sklearn.metrics import roc_auc_score
auc_score = roc_auc_score(y_test, y_proba)
print("Correct AUC (using probabilities):", auc_score*100)
```

```python
import numpy as np
k = int(0.05 * len(y_test))   # top 5% of test set
top_idx = np.argsort(y_proba)[-k:]   # indices of top-k highest probs
precision_at_k = y_test.iloc[top_idx].sum() / len(top_idx)
print("Precision@Top5%:", precision_at_k)
```

```python
model = XGBClassifier(n_estimators=200, max_depth=6, learning_rate=0.08, use_label_encoder=False, eval_metric='logloss', random_state=42)
model.fit(X_train, y_train, eval_set=[(X_test,y_test)], verbose=False)
```

```python
y_proba = model.predict_proba(X_test)[:,1]

```

```python
from sklearn.metrics import roc_auc_score, precision_recall_curve
auc = roc_auc_score(y_test, y_proba)
print("Auc_score",auc*100)
# precision@k: take top K% (or top K trips)
k = int(0.05 * len(y_test))   # top 5% as example
top_idx = np.argsort(y_proba)[-k:]
precision_at_k = y_test.iloc[top_idx].sum() / len(top_idx)
print("Precision@Top5%:", precision_at_k*100)
```
